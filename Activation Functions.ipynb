Sigmoid

```python
x = np.linspace(-10, 10, 100) 
z = 1/(1 + np.exp(-x)) 
  
plt.plot(x, z) 
plt.xlabel("x") 
plt.ylabel("Sigmoid(X)") 
  
plt.show() 
```

TanH

```python
import matplotlib.pyplot as plt 
import numpy as np 
import math 
  
x = np.linspace(-10, 10, 100) 
t=(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))  
plt.plot(x, t) 
plt.xlabel("x") 
plt.ylabel("tanh(X)") 
  
plt.show() 
```

ReLU

```python
def rectified(x):
    return max(0.0, x)
x = 1.0
print('rectified(%.1f) is %.1f' % (x, rectified(x)))
x = 1000.0
print('rectified(%.1f) is %.1f' % (x, rectified(x)))
x = 0.0
print('rectified(%.1f) is %.1f' % (x, rectified(x)))
x = -1.0
print('rectified(%.1f) is %.1f' % (x, rectified(x)))
x = -1000.0
print('rectified(%.1f) is %.1f' % (x, rectified(x)))
def rectified(x):
    return max(0.0, x)
series_in = [x for x in range(-10, 11)]
series_out = [rectified(x) for x in series_in]
plt.plot(series_in, series_out)
plt.show()
```

Leaky ReLU

```python
def leaky_relu(x):
  if x>0 :
    return x
  else :
    return 0.01*x

x = 1.0
print('Applying Leaky Relu on (%.1f) gives %.1f' % (x, leaky_relu(x)))
x = -10.0
print('Applying Leaky Relu on (%.1f) gives %.1f' % (x, leaky_relu(x)))
x = 0.0
print('Applying Leaky Relu on (%.1f) gives %.1f' % (x, leaky_relu(x)))
x = 15.0
print('Applying Leaky Relu on (%.1f) gives %.1f' % (x, leaky_relu(x)))
x = -20.0
print('Applying Leaky Relu on (%.1f) gives %.1f' % (x, leaky_relu(x)))
def rectified(x):
    if x>0 :
        return x
    else :
        return 0.01*x

series_in = [x for x in range(-10, 11)]
series_out = [rectified(x) for x in series_in]
plt.plot(series_in, series_out)
plt.show()
```

Exponential ReLU

```python
def exponential_relu(x):
  if x>0 :
    return x
  else :
    return 0.01*(np.exp(x)-1)

x = 1.0
print('Applying Exponential Relu on (%.1f) gives %.1f' % (x, exponential_relu(x)))
x = -10.0
print('Applying Exponential Relu on (%.1f) gives %.1f' % (x, exponential_relu(x)))
x = 0.0
print('Applying Exponential Relu on (%.1f) gives %.1f' % (x, exponential_relu(x)))
x = 15.0
print('Applying Exponential Relu on (%.1f) gives %.1f' % (x, exponential_relu(x)))
x = -20.0
print('Applying Exponential Relu on (%.1f) gives %.1f' % (x, exponential_relu(x)))
def rectified(x):
    if x>0 :
        return x
    else :
        return 0.01*(np.exp(x)-1)

series_in = [x for x in range(-10, 11)]
series_out = [rectified(x) for x in series_in]
plt.plot(series_in, series_out)
plt.show()
```
