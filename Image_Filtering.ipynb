```python
import numpy as np
def main():
    img = cv2.imread('img.jpg')
    plt.imshow(img, cmap=plt.cm.gray) 
    plt.axis('off')
    plt.show()
    marg = np.histogramdd(np.ravel(img), bins = 256)[0]/img.size
    marg = list(filter(lambda p: p > 0, np.ravel(marg)))
    entropy = -np.sum(np.multiply(marg, np.log2(marg)))
    print(entropy)
if __name__ == '__main__':
        main()
```

```python
import numpy as np
def plot_cv_img(input_image, output_image):
    fig, ax = plt.subplots(nrows=1, ncols=2)
    ax[0].imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
    ax[0].set_title('Input Image')
    ax[0].axis('off')
    ax[1].imshow(cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB))
    ax[1].set_title('Box Filter (3,3)')
    ax[1].axis('off')
    plt.show()
def main():
    img = cv2.imread('img.jpg')
    kernel_size = (3,3)
    blur = cv2.blur(img,kernel_size)
    plot_cv_img(img, blur)
    marg = np.histogramdd(np.ravel(blur), bins = 256)[0]/img.size
    marg = list(filter(lambda p: p > 0, np.ravel(marg)))
    entropy = -np.sum(np.multiply(marg, np.log2(marg)))
    print(entropy)
if __name__ == '__main__':
        main()
```
